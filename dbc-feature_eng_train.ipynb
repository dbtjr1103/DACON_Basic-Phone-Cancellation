```python
import pandas as pd
import random
import os
import numpy as np

from sklearn.cluster import KMeans
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import KFold
from sklearn.metrics import f1_score, make_scorer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier


def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42) # Seed 고정

train = pd.read_csv('/kaggle/input/dacon-basic-calls/train.csv')
test = pd.read_csv('/kaggle/input/dacon-basic-calls/test.csv')

# train 데이터를 train과 val로 나누기
train, val = train_test_split(train, test_size=0.2, stratify=train['전화해지여부'], random_state=42)

train = train[(train['가입일'] >= 1) & (train['가입일'] <= 800)]
train = train[(train['음성사서함이용'] >= 0) & (train['음성사서함이용'] <= 50)]
train.reset_index(drop=True, inplace=True)
val.reset_index(drop=True, inplace=True)

X = train.drop(['ID', '전화해지여부'], axis=1)
y = train['전화해지여부']

X_val = val.drop(['ID', '전화해지여부'], axis=1)
y_val = val['전화해지여부']
test = test.drop('ID', axis=1)

CLUSTERS = [5, 10]

for n_clusters in CLUSTERS:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    X[f'cluster_{n_clusters}'] = kmeans.predict(X)
    X_val[f'cluster_{n_clusters}'] = kmeans.predict(X_val)
    test[f'cluster_{n_clusters}'] = kmeans.predict(test)
```

```python
from sklearn.metrics import f1_score, make_scorer

# k-fold cross validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# macro f1 score를 사용하기 위해 make_scorer 함수를 사용하여 scorer 함수를 생성합니다.
scorer = make_scorer(f1_score, average='macro')

scores = []

models = []

for train_idx, val_idx in kf.split(X):
    x_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    x_val, y_val = X.iloc[val_idx], y.iloc[val_idx]
    
    # DecisionTreeClassifier 모델 학습
    model1 = ExtraTreesClassifier(random_state=42)
    model1.fit(x_train, y_train)
    models.append(('model1', model1))
    
    # RandomForestClassifier 모델 학습
    model2 = RandomForestClassifier(random_state=42)
    model2.fit(x_train, y_train)
    models.append(('model2', model2))
    
    # 검증 데이터 평가
    score1 = scorer(model1, x_val, y_val)
    score2 = scorer(model2, x_val, y_val)
    print(f"ExtraTreesClassifier score: {score1:.4f}, RandomForestClassifier score: {score2:.4f}")
    scores.append(score1)
    scores.append(score2)

# 평균 성능 계산
mean_score = np.mean(scores)
mean_score
```

```python
voting_model = VotingClassifier([(f'model_{i}', model[1]) for i, model in enumerate(models)], voting='soft')
voting_model.fit(X, y)

y_pred_val = voting_model.predict(X_val)
score_val = f1_score(y_val, y_pred_val, average='macro')
print(f"Validation F1 score: {score_val:.4f}")
```
